# Emotion recognition
### RNN-детектор когнитивной нагрузки

> Система, которая в фоновом режиме анализирует, как вы печатаете (паузы, зажатия клавиш, частоту ошибок), и предсказывает ваше состояние: _спокоен_ / _устал_ / _напряжён_ / _продуктивен_

### Вход

| Признак | Формула | Что означает | Эмоциональная связь |
|---------|---------|--------------|---------------------|
| **Hold Time** (HT) | `t_release - t_press` | Как долго зажата клавиша | Стресс → дольше держим |
| **Flight1** (F1) | `t(i+1, press) - t(i, release)` | Пауза между отпусканием и следующим нажатием | Усталость → паузы нестабильны |
| **Flight2** (F2) | `t(i+1, release) - t(i, release)` | От отпускания до отпускания следующей клавиши | Характеризует ритм |
| **Flight3** (F3) | `t(i+1, press) - t(i, press)` | От нажатия до нажатия | «Сырой» межклавишный интервал |
| **Flight4** (F4) | `t(i+1, release) - t(i, press)` | Полное время на две клавиши | Используют в аутентификации |

Дополнительно (сильный сигнал):
- `is_backspace`: бинарный флаг (1 если это Backspace или Delete).
- `is_punctuation`: бинарный флаг (усталость → пауза после точки/запятой увеличивается)
- `hour_of_day`: час (circadian ритмы влияют на когнитивную нагрузку)

Вектор для одного нажатия: 
```py
[HT, F1, F2, F3, F4, is_backspace, is_punctuation, hour_of_day]
```

### Датасет

> [EmoSurv](https://ieee-dataport.org/open-access/emosurv-typing-biometric-keystroke-dynamics-dataset-emotion-labels-created-using)

- Набор, содержащий данные о нажатиях клавиш вместе с метками эмоций.
- Данные о времени и частоте нажатия записываются, когда участники печатают произвольные и фиксированные тексты до и после того, как у них возникли определенные эмоции. 
- Возможные эмоции: счастье (`H`), спокойствие (`C`), грусть (`S`), нейтральное состояние (`N`)
